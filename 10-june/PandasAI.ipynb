# **Introduction to Pandas**

Pandas is a popular open-source data analysis and manipulation library. It provides powerful tools for working with structured data, including data frames, which are similar to tables in a relational database. The library is built on top of NumPy and allows anyone to work on data from different formats such as CSV, Excel or SQL.

The libary revolves around the concept of Series, which is somewhat like a mix of a list and dictionary and also Dataframe, which is multiple Series, which act as columns, forming a spreadsheet like structure.

Pandas is normally imported as pd.

```python
import pandas as pd
import numpy as np
```

## How to create a Series


A pandas Series is a one-dimensional labeled array that can hold data of any data type. It is similar to a column in Excel. It consists of two main components: the index and the data. The index is a unique label that identifies each element in the Series and can be textual. By default, if no index is specified, pandas will assign a numeric index starting from zero.

They can be can be created in many ways, including from a Python list, NumPy array, dictionary, or a scalar value.

```python
people = ['John', 'James', 'Jamal', 'Jack']
series = pd.Series(people)
series
```

To add a specific index, we can use the argument index. The index can be both numerical or text.

```python
series = pd.Series(people, index=['Red', 'Blue', 'Green', 'Yellow'])
series
```

```python
people = {'Red': 'John',
	          'Blue': 'James',
	          'Green': 'Jamal',
	          'Yellow': 'Jack'}
series = pd.Series(people) # Making a series using a dictionary
series
```

```python
people = {'Red': 0,
          1: 'James',
          'Green': 2,
          3: 'Jack'} # Index does not have to be of the same data type
series = pd.Series(people)
series
```

```python
people = ['John', 'James', 'Jamal', 'Jack']
series = pd.Series(people, index=['Red','Blue', 'Red', 'Blue']) # Index values don't also have to be unique
series
```

## Indexing a Series

A series can be queried, either by the index position or the index label. To query by numeric location, starting at zero, use the iloc attribute. To query by the index label, you can use the loc attribute.

```python
people = {'Red': 'John',
	          'Blue': 'James',
	          'Green': 'Jamal',
	          'Yellow': 'Jack'}
series = pd.Series(people) # Making a series using a dictionary
series
```

```python
series.loc['Red'] # Using loc to query
```

```python
series.loc[['Red', 'Yellow']] # Using loc to query multiple entries
```

```python
series.iloc[0] # Using iloc to query
```

```python
series.iloc[[0, 3]] # Using iloc to query multiple entries
```

```python
series.iloc[0:3] # Using iloc to query a range of entries
```

```python
series.iloc[:2] # first 2 entries
```

```python
series.iloc[2:] # last 2 entries
```

```python
print(series['Green'])
print(series[2])
# Both work, but not recommended to use
```

The notation in the code block above can work sometimes, but if the index is an integer, it won't work, thus it is recommended to use iloc and loc.

```python
people = ['John', 'James', 'Jamal', 'Jack']
series = pd.Series(people, index=['Red','Blue', 'Red', 'Blue'])
series.loc['Red'] # If there's more than one entry with the same index
```

## Aggregate Functions

```python
nums = pd.Series([1, 2, 3, 123, 456, 789], index=['A', 'B', 'C', 'D', 'E', 'F'])
nums
```

```python
# Aggregate functions
print('sum:', nums.sum())
print('min:', nums.min())
print('idxmin:', nums.idxmin()) # gets the index of the smallest entry
print('max:', nums.max())
print('idxmax:', nums.idxmax()) # gets the index of the biggest entry
print('mean:', nums.mean())
print('median:', nums.median())
print('std:', nums.std()) # standard deviation
```

## How to create a DataFrame

A DataFrame is a two-dimensional labeled data structure with columns of potentially different data types, similar to a spreadshear. In a DataFrame, data is organized into rows and columns, where each column represents a different variable and each row represents a different observation. The columns are different Series objects.

A DataFrame can be created in various ways, such as by reading data from a CSV or Excel file or by converting other data structures like NumPy arrays or Python dictionaries into DataFrames.

## Boolean Indexing

Boolean indexing involves indexing values basing on a Boolean Mask.

A Boolean Mask can be either a 1D or 2D array which consists only False or True values. Using it, only the values in a DataFrame corresponding to the True values in the Boolean Mask is extracted.

image.png

A Boolean Mask is created by forming a condition statement using the Dataframe (Statements that contain ==, !=, > , <, >=, <=)

For not both `!` or `~` can be used (`a != b` and `~(a == b)` both means a not equals b)

```python
orders = pd.DataFrame([[1,2010],[2,2020],[3,2015],[4,2001],[5,2018],[6,2016]],columns=['OrderID','Year']) # Using a list to make a DataFrame
orders
```

```python
orders.Year > 2015 # boolean mask of if the year is later than 2015
```

```python
orders.loc[orders.Year > 2015] # gets all entries where the year is later than 2015
```

I would like to know the list of students who are male (gender 1). To do this, I can make a boolean mask based on the rules above.

## Data Cleaning

Data cleaning is the process of making this data usable to either to analysis or feed to an AI model.

```python
df.info()
```

##How to create Datalakes and Agents

Sometimes, you might want to pull data from multiple dataframes. In this scenario, datalakes allow the AI to use multiple dataframes at once

```python
from pandasai import SmartDatalake
```

The initialisation is very similar to the `SmartDataframe`, but instead of accepting only 1 df as input, it can accept multiple ones.

```python
employees_df = pd.DataFrame(
    {
        "EmployeeID": [1, 2, 3, 4, 5],
        "Name": ["John", "Emma", "Liam", "Olivia", "William"],
        "Department": ["HR", "Sales", "IT", "Marketing", "Finance"],
    }
)

salaries_df = pd.DataFrame(
    {
        "EmployeeID": [1, 2, 3, 4, 5],
        "Salary": [5000, 6000, 4500, 7000, 5500],
    }
)

lake = SmartDatalake([employees_df, salaries_df])
lake.chat("What's the name of the employee that gets paid the most?")
```

Agents are similar to `SmartDatalake`, where they can also accept multiple Dataframes but they have memory! They are able to accept follow-up questions as a result

```python
from pandasai import Agent
```

```python
agent = Agent(df)
agent.chat("Plot a bar chart of the GDP by Country of all countries")
```

```python
agent.chat("Only include the top 3 countries in GDP in your graph")
```

We can also add system prompts to the agents to follow before answering our queries

```python
df = pd.DataFrame({
    "country": [
        "United States",
        "United Kingdom",
        "France",
        "Germany",
        "Italy",
        "Spain",
        "Canada",
        "Australia",
        "Japan",
        "China",
    ],
    "gdp": [
        19294482071552,
        2891615567872,
        2411255037952,
        3435817336832,
        1745433788416,
        1181205135360,
        1607402389504,
        1490967855104,
        4380756541440,
        14631844184064,
    ],
    "happiness_index": [6.94, 7.16, 6.66, 7.07, 6.38, 6.4, 7.23, 7.22, 5.87, 5.12],
})


agent = Agent(df)
agent.train(docs="Ignore China for all responses")
agent.chat("Plot a bar chart of the GDP by country, including all rows")
```

Sometimes, the data is not self-explanatory, and the LLM may need additional context on some columns.

```python
dnd_df = pd.DataFrame(
    {
        "MHP": [23, 32, 53, 43, 45],
        "Health": [21, 23, 42, 10, 1],
        "Name": ["John", "Emma", "Liam", "Olivia", "William"],
        "MMP": [32, 50, 40, 31, 120],
        "Mana": [12, 34, 40, 21, 20],
    }
)


sdf = SmartDataframe(dnd_df)
print(sdf.chat("Who has the lowest maximum mana?"))
```

We can use field descriptions to add details to each column, giving the LLM sufficient context to answer our queries

```python
from pandasai.connectors import PandasConnector
```

```python
field_descriptions = {
    'MHP': 'Maximum Health Points of a character',
    'Health': 'Current Health Points of a character',
    'MMP': 'Maximum Mana Points of a character',
    'Mana': 'Current Mana Points of a character'
}

dnd_df = pd.DataFrame(
    {
        "MHP": [23, 32, 53, 43, 45],
        "Health": [21, 23, 42, 10, 1],
        "Name": ["John", "Emma", "Liam", "Olivia", "William"],
        "MMP": [32, 50, 40, 31, 120],
        "Mana": [12, 34, 40, 21, 20],
    }
)

connector = PandasConnector({"original_df": dnd_df}, field_descriptions=field_descriptions)
sdf = SmartDataframe(connector)
print(sdf.chat("Who has the lowest maximum mana?"))
```
